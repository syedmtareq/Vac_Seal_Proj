```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt

headers = ['Wave','Test_1','Test_2','Test_3','Test_4','Test_5_Without','Test_6_Without','Test_7_Without']

# df = pd.read_csv(r':\Users\user\Desktop\Vac_Seal\UVvisNIR_Arbitrary_4.csv')

df = pd.read_csv('r'C:\Users\user\Desktop\Vac_Seal\UVvisNIR_Arbitrary_4.csv', names=headers, skiprows=[0])

plt.plot(df['Wave'], df['Test_1'], '-',color='red', label='No coating', lw=1)
plt.plot(df['wave'], df['Test_2'], '-', color='orange', label='10 cycles', lw=1)
plt.plot(df['wave'], df['Test_3'], '-', color='blue', label='25 Cycles', lw=1)
plt.plot(df['wave'], df['Test_4'], '-', color='green', label='50 Cycles', lw=1)
plt.plot(df['wave'], df['Test_5_Without'], '-', color='purple', label='100 Cycles', lw=1)
plt.plot(df['wave'], df['Test_6_Without'], '--', color='purple', label='100 Cycles', lw=1)
plt.plot(df['wave'], df['Test_7_Without'], '-.', color='purple', label='100 Cycles', lw=1)

plt.xlabel('Wavelength ($nm$)')
plt.ylabel('Transmittance (%)')
plt.xticks(np.arange(0, 2501, 500))
plt.xlim([250, 2500])
plt.ylim([50, 102])
plt.legend(loc='best', bbox_to_anchor=(0.03, 0.4))
plt.savefig('trans_ald_effect.jpeg', format='jpeg', dpi=1000)
plt.show()
```
