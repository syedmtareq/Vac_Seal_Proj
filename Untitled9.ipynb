```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

df = pd.read_csv('D:\Yang\Practice\ArbitrarySample_2\\Specimen_RawData_5.csv')
df
```

```python
from scipy.signal import lfilter

plt.rcParams["figure.figsize"] = (16,10)

# plt.plot(df.Extension, df['16 D_without seal'])
n = 100  # the larger n is, the smoother curve will be
b = [1.0 / n] * n
a = 1
yy = lfilter(b,a, df['7 D_1_with'])
#plt.plot(df.Extension, yy, linewidth=2, linestyle="-", color="blue")7 D_1_with
plt.plot(df.Extension, yy, linewidth=2, linestyle='-', color="blue", label='7 D_1_with')



# plt.plot(df.Extension, df['16 D_with seal'])
n = 100  # the larger n is, the smoother curve will be
b = [1.0 / n] * n
a = 1
yy = lfilter(b,a, df['7 D_1_without'])
#plt.plot(df.Extension, yy, linewidth=2, linestyle="-", color="red")
plt.plot(df.Extension, yy, linewidth=2, linestyle="dashed", color="blue", label='7 D_1_without' )


# plt.plot(df.Extension, df['21 D_without seal'])
n = 100  # the larger n is, the smoother curve will be
b = [1.0 / n] * n
a = 1
yy = lfilter(b,a, df['7 D_2_with'])
plt.plot(df.Extension, yy, linewidth=2, linestyle='-', color="orange",label='7 D_2_with' )





# plt.plot(df.Extension, df['21 D_with seal'])
n = 100  # the larger n is, the smoother curve will be
b = [1.0 / n] * n
a = 1
yy = lfilter(b,a, df['7 D_2_without'])
plt.plot(df.Extension, yy, linewidth=2, linestyle="dashed", color="orange", label='7 D_2_without' )
plt.legend(loc='best',fontsize=14)
plt.title('Load vs Extension', fontdict={'fontsize': 16})
plt.xlabel('Extension (mm)',fontdict={'fontsize': 14})
plt.ylabel('Load(MPa)',fontdict={'fontsize': 14})

# plt.legend()

# plt.plot(strain_per, stress, '-', color = 'g',linestyle='dashed', label='21 Days_without sealing')
# plt.legend(loc='best',fontsize=14)
# plt.title('Aging Strengths', fontdict={'fontsize': 16})
# plt.xlabel('% Strain',fontdict={'fontsize': 14})
# plt.ylabel('Flexural Stress (MPa)',fontdict={'fontsize': 14})

plt.show()

```
